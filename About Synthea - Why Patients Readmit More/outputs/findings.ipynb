# ***Findings ->***

- ***`URGENTCARE` AND `AMBULATORY` `ENCOUNTER CLASS` HAVE THE `MOST CHANCE OF ANOTHER READMISSION THAN OTHERS`***

- ***`MARITAL STATUS` `SINGLE` HAVE `HIGHER CHANCE OF CAUSING ANOTHER READMISSION`***

- ***PATIENTS WITH `BLACK` AND `NATIVE` `RACE` HAVE A `HIGH CHANCE OF CAUSING ANOTHER READMISSION`***

- ***`MALE` `PATIENTS` HAVE SLIGHTLY `MORE CHANCE OF CAUSING A READMISSION`***

- ***`FEMALE` `PROVIDERS` TREATING AN ENCOUNTER `HAVE MORE CHANCE OF CAUSING A READMISSION`***

- ***`ENCOUNTER DURATION` MORE THAN `30 MINUTES` HAVE `MORE CHANCE OF CAUSING READMISSION`***

- ***`PROCEDURE COST` `ABOVE 1610` HAVE A `HIGH CHANCE OF CAUSING ANOTHER READMISSION`***

- ***`LOW(1-7)` AND `HIGH(FROM 264)` `BASE MEDICINE COST` HAVE `HIGHER CHANCE OF ANOTHER READMISSION, COMPARED TO MEDIUM MEDICINE COST`***

- ***`TOTAL MEDICINE COST` `UNDER 727` HAVE `HIGH CHANCE OF CAUSING ANOTHER READMISSION`***

- ***`AGE` GROUPS `19-50`, `59-65` AND `80+` HAVE `HIGH CHANCES OF CAUSING ANOTHER READMISSION`***

- ***`HIGHER` THE `UTILIZATION`, MORE THE `REASON FOR CAUSING ANOTHER READMISSION`***

- ***`COVERAGE RATIO` `LESS THAN 70%` MORE PROBABILY `REASON TO ANOTHER READMISSION`***

- ***`MEDICATION DURATION` `LESS THAN 3 MONTHS` IS MORE LIKELY TO `BE IN ANOTHER READMISSION`***
