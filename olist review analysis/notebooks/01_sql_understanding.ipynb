### **So for our REVIEW ANALYSIS project we do not need all the values and all the tables**

**what we need is ->**

- from `customers` table - the `city` of the customer

- from `order_items` table - how many `product_id`, `total price`, `total freight`, if used multiple `seller_id`

- from `order_payments` table - how many `sequence_payment` used,how many `payment_type` used,how many `installments` used, total paid `payment_value`

- we will use `reviews` table as main table, but only use `review_id`, `order_id`, `review_score`, `review_creation_date` columns

- from `order` table - what is the `status` ,`approval_delay`,`delivery_days`, how many days late delivery

**special scenario, we need to merge orders and products table, fetch total weight,length,height,width to calculate some measurement in each order,and how many categories were ordered in that order**

- from `seller` table - the `city` of the seller

### **Creating backup tables**

```python
df = {
    'olist_customers_dataset' : 'customers',
    'olist_order_items_dataset' : 'order_items',
    'olist_order_payments_dataset': 'order_payments',
    'olist_order_reviews_dataset' : 'order_reviews',
    'olist_orders_dataset' : 'orders',
    'olist_products_dataset' : 'products',
    'olist_sellers_dataset' : 'sellers'
}
```

```python
for table in df.items():
    print(table[0],table[1])
```

```python
for table in df.items():
    query = f'''
        create table {table[1]} like {table[0]}
    '''
    F.run(query)
    query = f'''
    insert into {table[1]} select * from {table[0]}
    '''
    F.run(query)
```
