### Understanding about this data


After checking the datasets, I can understand the encounters table is the major table where each patient visit is recorded
- and for our analysis of patient readmission, we do not need to use all the tables

- some important tables are - `encounters`,`allergies`,`careplan`,`conditions`,`medications`,`procedures`,`patients`,`providers`

- I will not use `organization` or `payer` table, as they are not directly related to `patients` and `encounters`

- need to clean these tables, with keeping backup...and create a new dataframe by using aggregated column according to my need

```python
necessary_tables = ['encounters','allergies','careplans',
                    'conditions','medications',
                    'procedures','patients','providers']
```

```python
import time


start = time.time()
for table in necessary_tables:
    query = f'''
        create table backup_{table} like {table}
    '''
    F.run(query)
    query = f'''
    insert into backup_{table} select * from {table}
    '''
    F.run(query)
end = time.time()
print(f"Time Taken: {round((end - start),2)}")
```
